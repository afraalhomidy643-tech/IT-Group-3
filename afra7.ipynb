```python
import pandas as pd


data = {
    'الاسم': ['أحمد', 'فاطمة', 'محمد', 'سارة', 'خالد'],
    'العمر': [25, 30, 22, 28, 35],
    'المدينة': ['صنعاء', 'عدن', 'تعز', 'إب', 'الحديدة'],
    'الدرجة': [85, 90, 78, 92, 88]
}

```

```python
df = pd.DataFrame(data)

print("عينة من البيانات:")
display(df.head())
```

```python

print("معلومات البيانات:")
df.info()
```

```python
print("الإحصائيات الوصفية:")
display(df.describe())
```

```python

df_filtered = df[df['الدرجة'] > 80]


print("عدد الصفوف والأعمدة بعد التصفية:")
print(df_filtered.shape)
```

```python
print("البيانات بعد التصفية:")
display(df_filtered)
```
